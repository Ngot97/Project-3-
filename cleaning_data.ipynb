```python
from pathlib import Path
import pandas as pd
```

```python
books_path = Path("Resources/books.csv")

books_df = pd.read_csv(books_path, encoding="utf-8")
books_df.head()
```

```python
users_path = Path("Resources/users.csv")

users_df = pd.read_csv(users_path, encoding="utf-8")
users_df.head()
```

```python
ratings_path = Path("Resources/ratings.csv")

ratings_df = pd.read_csv(ratings_path, encoding="utf-8")
ratings_df.head()
```

```python
#remove url columns
drop_columns = ["Image-URL-S",
               "Image-URL-M",
               "Image-URL-L"]

books_df.drop(drop_columns, inplace=True, axis=1)
books_df.head()
```

```python
rename_columns = {"Book-Title": "Book_Title",
                 "Book-Author": "Book_Author",
                 "Year-Of-Publication": "Year_Of_Publication",
                 }

books_df.rename(columns=rename_columns, inplace=True)

books_df.head(55)
```

```python
books_df.dtypes
```

```python
# Your existing code for conversion to numeric and dropping NaN
books_df["Year_Of_Publication"] = pd.to_numeric(books_df["Year_Of_Publication"], errors="coerce")
books_df = books_df.dropna(subset=["Year_Of_Publication"])

# Filter based on numeric values
new_books_df = books_df[pd.notna(books_df["Year_Of_Publication"])]

# Convert ‘Year_Of_Publication’ to int
new_books_df["Year_Of_Publication"] = new_books_df["Year_Of_Publication"].astype(int)
new_books_df.dtypes
```

```python
new_names_users = {"User-ID": "User_ID"}

users_df.rename(columns=new_names_users, inplace=True)

users_df.head()
```

```python
users_df.dtypes
```

```python
# split on location
split_location = users_df["Location"].str.split(", ", expand=True)

# Check if there are three elements after splitting
if len(split_location.columns) == 3:
    users_df[["City", "State", "Country"]] = split_location
else:
    # Display rows where the split result doesn’t have three elements
    invalid_rows = split_location[split_location.apply(lambda row: len(row.dropna()) != 3, axis=1)]
    if not invalid_rows.empty:
        print("Rows with incomplete split:")
        print(invalid_rows)
        print("\n")
        
        # Examine and analyze the cases where the split result doesn’t have three elements
        for index, row in invalid_rows.iterrows():
            print(f"Row {index + 1}: {row}")
            
        print("\n")
        
        # Create a new DataFrame with only the invalid rows
        invalid_df = users_df.loc[invalid_rows.index]
        # Remove invalid rows from the original DataFrame
        users_df = users_df.drop(invalid_df.index)
        print("Removed invalid rows from the original DataFrame.")
        
    else:
        print("No rows with incomplete split found.")

# Now handle the case where the split result doesn’t have three elements
if len(split_location.columns) != 3:
    print("Error: Unable to split Location into three columns.")
```
